1. Dataset Size: 100 speakers * 200 recordings/day * 5 days ≈ 100,000 phone recordings.

```python
# 1-NN Classification:
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.neighbors import KNeighborsClassifier

# Positive (red points)
# Negative (blue points)
X = np.array([[1, 2], [1, 4], [5, 4], [3, 1], [3, 2]])
y = np.array([1, 1, 1, 0, 0])  # 1 = Positive, 0 = Negative

# Train 1-NN classifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X, y)

# Generate grid points for decision boundary
x_min, x_max = 0, 6
y_min, y_max = 0, 5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                     np.linspace(y_min, y_max, 300))

# Predict class for each grid point
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

cmap_light = ListedColormap(['#AAAAFF', '#FFAAAA']) # Decision regions
cmap_bold = ListedColormap(['#0000FF', '#FF0000']) # Actual data points
plt.figure(figsize=(6, 5))
plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.6)

# Training samples
plt.scatter(X[y == 1, 0], X[y == 1, 1],
            c='red', label='Positive', edgecolor='k', s=100)
plt.scatter(X[y == 0, 0], X[y == 0, 1],
            c='blue', label='Negative', edgecolor='k', s=100)
plt.xlabel('Feature 1 (x)')
plt.ylabel('Feature 2 (y)')
plt.title('1-NN Decision Boundary (1-NN Decision Boundary)')
plt.legend()
plt.grid(True)
plt.show()
```

2. **Before scaling:**
   The nearest neighbor of (500, 1) is (500, 4), with a distance of 3.
   Since (500, 4) is a Positive sample, the 1-NN classifier assigns Positive to (500, 1).

   **After scaling to [0, 1]:**
   Using min-max normalization for each feature (x: min=100, max=500; y: min=1, max=4)

   After scaling, (500, 1) becomes (1, 0).

   The nearest neighbor is now (300, 1) → (0.5, 0), with a distance of 0.5.

   Since (300, 1) is a Negative sample, the 1-NN classifier assigns Negative to (500, 1).
   
3. **Handling Missing Values**

   Use available features only: When computing distances, ignore the missing features and calculate Euclidean or other distances based only on the available ones.

   Impute missing values: Replace missing features with the mean, median, or mode of that feature in the training data before computing distances.

   Use weighted distance metrics: Apply distance measures such as Gower distance, which assign a weight of 0 to missing features so that only present features contribute to the overall distance.

4. **High-dimensional Data**

   Low-dimensional manifold: Image data often lie on a low-dimensional manifold embedded in high-dimensional space. Similar images cluster together along this manifold.

   Correlated features: Pixels are highly correlated and structured, so distance metrics such as Euclidean distance or cosine similarity can effectively capture visual similarity.

   Feature extraction: In practice, feature extraction methods (e.g., SIFT, HOG, or deep neural network embeddings) reduce the effective dimensionality and make K-NN more robust and efficient.

### Problem 3: Two-point 2D Dataset (Part 2)

Run the Perceptron algorithm **by hand or in code** on the following data:

1. Positive class: (10, -2)
2. Negative class: (12, 2)

Start with $w_0 = (0, 0)$ and a learning rate of 1.

- Compute how many updates are required until convergence.
- Write down the sequence of $w_i$ vectors.

```python
# Todo
import numpy as np

X = np.array([[10, -2], # Positive class
              [12,  2]]) # Negative class
y = np.array([1, -1]) # Corresponding labels

# Initialize parameters
w = np.array([0.0, 0.0]) # Initial weight vector w0 = (0,0)
eta = 1 # Learning rate
updates = 0 # Update counter
history = [w.copy()] # Record sequence of w

# Run Perceptron updates until convergence
converged = False
while not converged:
    converged = True
    for i in range(len(X)):
        # Check classification correctness
        if y[i] * np.dot(w, X[i]) <= 0:
            # If misclassified → update 
            w = w + eta * y[i] * X[i]
            updates += 1
            history.append(w.copy())
            converged = False

print(f"Total updates until convergence: {updates}")
print("Sequence of weight vectors:")
for idx, vec in enumerate(history):
    print(f"w_{idx} = {vec}")
```

### Problem 4: Reconstructing the Weight Vector

Given the log of Perceptron updates:

| x | y | count |
|---|---|--------|
| (0, 0, 0, 0, 4) | +1 | 2 |
| (0, 0, 6, 5, 0) | +1 | 1 |
| (3, 0, 0, 0, 0) | -1 | 1 |
| (0, 9, 3, 6, 0) | -1 | 1 |
| (0, 1, 0, 2, 5) | -1 | 1 |

Assume learning rate = 1 and initial weight $w_0 = (0, 0, 0, 0, 0)$.

Compute the final weight vector after all updates.

```python
#Todo
import numpy as np
# (x, y, count)
data = [
    (np.array([0, 0, 0, 0, 4]),  +1, 2),
    (np.array([0, 0, 6, 5, 0]),  +1, 1),
    (np.array([3, 0, 0, 0, 0]),  -1, 1),
    (np.array([0, 9, 3, 6, 0]),  -1, 1),
    (np.array([0, 1, 0, 2, 5]),  -1, 1)
]

# w0 = (0,0,0,0,0)
w = np.zeros(5)

# y * x * count
for x, y, count in data:
    w += y * x * count

print("Final weight vector:", w)
```

### Problem 5: Visualizing Perceptron Convergence

Implement a Perceptron on a small 2D dataset with positive and negative examples.

- Plot the data points.
- After each update, visualize the decision boundary.
- Show how it converges to a stable separator.

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time

X = np.array([
    [2, 2],
    [3, 1],
    [2, 1.5],
    [-2, -1],
    [-3, -2],
    [-1.5, -2.5]
])
y = np.array([1, 1, 1, -1, -1, -1])

# Bias term
Xb = np.hstack([X, np.ones((X.shape[0], 1))])  # shape = (N, 3)
w = np.zeros(3)  # Initial weights (w1, w2, b)
eta = 1.0

def plot_boundary(w, X, y, title=""):
    plt.figure(figsize=(5, 5))
    plt.scatter(X[y == 1, 0], X[y == 1, 1], label="Positive (+1)")
    plt.scatter(X[y == -1, 0], X[y == -1, 1], label="Negative (-1)")

    # y = -(w1*x + b)/w2
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    if abs(w[1]) > 1e-6:
        xs = np.linspace(x_min, x_max, 100)
        ys = -(w[0] * xs + w[2]) / w[1]
        plt.plot(xs, ys, 'k-', linewidth=2)
    plt.xlim(x_min, x_max)
    plt.ylim(X[:, 1].min() - 1, X[:, 1].max() + 1)
    plt.title(title)
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.legend()
    plt.grid(True)
    plt.show()

max_epochs = 20
updates = 0

for epoch in range(max_epochs):
    error_count = 0
    for i in range(len(Xb)):
        if y[i] * np.dot(w, Xb[i]) <= 0:
            w = w + eta * y[i] * Xb[i]
            updates += 1
            error_count += 1
            clear_output(wait=True)
            plot_boundary(w, X, y, title=f"Update {updates}: w = {w}")
            time.sleep(0.8)
    if error_count == 0:
        break

print(f"Converged after {updates} updates. Final w = {w}")
```
